```python
import cv2
import os

data_dir = '/home/yaosy/Diskb/bdmeet/mowei/mowei_squareMov'
listd = sorted(os.listdir(data_dir))
```

```python
def crop(Ai):
    w = Ai.shape[0]
    base = 32
    x_cen = w // 2
    bs = int(w * 0.25) // base * base
    return Ai[:,(x_cen-bs):(x_cen+bs),:]
```

```python
fps = 30
size = (512, 512)
videowriter= cv2.VideoWriter('mowei_30fps512_crop.avi', cv2.VideoWriter_fourcc(*'XVID'), fps, (256, 512))
for i,name in enumerate(listd):
#     num = listd[i].split('_')[1][:4]
#     name = 'merge_{}.png'.format(num)
#     if i > 50:
#         break
    print(i)
    print(os.path.join(data_dir, name))
    img = cv2.imread(os.path.join(data_dir, name))
    img = cv2.resize(img, size)
    img = crop(img)
    print(img.shape)
    videowriter.write(img)
```
